```python
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
```

```python
output_path = working_path = "/home/acucena/Documentos/UFPI/ICV/Resultado/"

g = glob(output_path + 'maskedimages_*.npy')
imgs = np.load(g[3])
```

```python
img = imgs[162][0]
imgf = img.astype(np.float64)
```

```python
plt.imshow(imgf, cmap='gray')
plt.show()
```

```python
for segs in (100, 200, 300, 400):
    segments = slic(imgf, n_segments=segs, sigma=5)
    plt.imshow(segments)
    plt.show()
```

```python
for segs in (100, 200, 300, 400):
    segments = slic(imgf, n_segments=segs, compactness=10, sigma=1)
    plt.imshow(mark_boundaries(imgf,segments))
    plt.show()
```
